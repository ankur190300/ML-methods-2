# Data Augmentation
- Most of the neural networks are "data hungry"
- This leads to overfitting of the neural network
- To avoid overfitting data can be augmented and fed to the network


```python
from keras.preprocessing import image
import numpy as np
```

```python
path='./images/cbsupe.jpeg'
img= image.load_img(path, target_size=(100,100))
```

```python
import matplotlib.pyplot as plt
x= image.img_to_array(img)
x/= 255.0
x=x.reshape((1,100,100,3))
print(x.shape)
```

```python
 def drawimg(X, title = "pic") :  
    plt.style.use('seaborn')
    plt.axis('off')
    plt.title(title)
    plt.imshow(x)
```

```python
datagen = image.ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    vertical_flip=True,
    )
i=0
for batch in datagen.flow(x,batch_size=1):
    plt.figure()
    imgplot= plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if(i==10):
        break
    plt.show()
    
```
