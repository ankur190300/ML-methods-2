```python
import cv2
import matplotlib.pyplot as plt
```

```python
im= cv2.imread('./images/elephant.jpeg') #in BGR format
print(im.shape)
```

```python
plt.imshow(im)
```

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
```

```python
# Flatten the image to a 2-D array
pixels= im.reshape((-1,3))
print(pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors= 4 # number of dominant colours we have to find in a picture
kmeans = KMeans(n_clusters= dominant_colors)
kmeans.fit(pixels)
```
