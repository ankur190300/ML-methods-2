```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, Y= make_blobs(n_samples=500,
    n_features=2,
    centers=5,
    random_state=3,)
```

```python
kmeans= KMeans(n_clusters=5)
kmeans.fit(X)
```

```python
centers=kmeans.cluster_centers_
print(centers)
```

```python
labels= kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1], marker="*", color= "orange")
plt.style.use("default")
#plt.title("prediction")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(centers[:,0],centers[:,1], marker="*", c= "black")
#plt.style.use("seaborn")
plt.title("original data")
plt.show()
```

- K-means ++ however can only sperate data lenearly and thus can't be used everywhere

```python
from sklearn.datasets import make_circles, make_moons
```

```python
X2, Y2= make_moons(n_samples=500, shuffle=True, noise=0.1, random_state=3)
```

```python
plt.scatter(X2[:,0],X2[:,1],c=Y2)
```

```python
kmeans2= KMeans(n_clusters=2)
kmeans2.fit(X2)
centers2=kmeans2.cluster_centers_
labels2= kmeans2.labels_
```

```python
plt.scatter(X2[:,0],X2[:,1],c=labels2)
plt.scatter(centers2[:,0],centers2[:,1], marker="*", color= "orange")
plt.style.use("default")
#plt.title("prediction")
plt.show()
```

```python
plt.scatter(X2[:,0],X2[:,1],c=Y2)
plt.scatter(centers2[:,0],centers2[:,1], marker="*", c= "black")
#plt.style.use("seaborn")
plt.title("original data")
plt.show()
```

- So here the k-means ++ classification failed , hence we use DBSCAN

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbs= DBSCAN(eps=0.2 , min_samples=8 )
dbs.fit(X2)
```

```python
y= dbs.fit_predict(X2)
print(y)
```

```python
plt.scatter(X2[:,0],X2[:,1],c= y)
```

- As we can see DBSCAN works better
